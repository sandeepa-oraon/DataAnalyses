```python
import pandas as pd
```

#Sorting

Sort DataFrame by One Column Value

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 95, 80]}
df = pd.DataFrame(data)
```

using sort_values()

```python
#in ascending order
df.sort_values(by='Age')
```

```python
#in descending order(set ascending to 'False')
df.sort_values(by='Age', ascending=False)
```

##Sort DataFrame by Multiple Columns

```python
df.sort_values(by=['Age', 'Score'])
```

This will first sort by Age, and if there are ties (same age), it will then sort by Score.

##Sort DataFrame with Missing Values

```python
data_with_nan = {"Name": ["Alice", "Bob", "Charlie", "David"],"Age": [28, 22, None, 22]}
df_nan = pd.DataFrame(data_with_nan)
```

```python
df_nan
```

```python
df_nan.sort_values(by= 'Age', na_position='first')
```
