```python
data=[['Stu1', 28, 'Engineer'],
      ['Stu2', 25, 'Data Scientist'],
      ['Stu3', '32', 'Manager']]
```

```python
columns= ['Name', 'Age', 'Occupation']
df= pd.DataFrame(data, columns=columns)
df['Age']= pd.to_numeric(df['Age'])
```

```python
data = [['Geek1', 28, 'Engineer'],
        ['Geek2', None, 'Data Scientist'],
        ['Geek3', 32, None]]
column= ['Name', 'Age', 'Occupation']
df= pd.DataFrame(data, columns=column)
df
```

```python
df= df.replace({None: np.nan})
df
```

creating DataFrame from Multiple Series

```python
author = ['Jitender', 'Purnima',
          'Arpit', 'Jyoti']
article = [210, 211, 114, 178]

aut_ser= pd.Series(author)
aut_art= pd.Series(article)

frame={'Author': aut_ser,
       'Article': aut_art }
df= pd.DataFrame(frame)

df
```

```python
age= [21, 23, 21, 20]
df['Age']= pd.Series(age)
df
```
