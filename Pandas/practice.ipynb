#Convert list of nested dictionary into Pandas dataframe

```python
countries = {
    "1": {"Country": "India",
          "Capital": "Delhi",
          "Population": "5,123,456,789"},
    "2": {"Country": "New Country 2",
          "Capital": "New Capital 2",
          "Population": "987,654,321"},
    "3": {"Country": "New Country 3",
          "Capital": "New Capital 3",
          "Population": "111,222,333"}
}
```

```python
#from_dict() gives the more control over key(whether rows and columns)
df= pd.DataFrame.from_dict(countries)
df
```

```python
df= pd.DataFrame.from_dict(countries, orient='index')
df
```

##Nested Dictionary to Pandas DataFrame using Native Method

```python
li = [{
        "Student": [{"Exam": 90, "Grade": "a"},
                    {"Exam": 99, "Grade": "b"},
                    {"Exam": 97, "Grade": "c"},
                    ],
        "Name": "Paras Jain"
        },
        {
        "Student": [{"Exam": 89, "Grade": "a"},
                    {"Exam": 80, "Grade": "b"}
                    ],
        "Name": "Chunky Pandey"
        }]
```

```python
li
```

```python
rows= []

for data in li:
  rData= data['Student']
  name= data['Name']

  for row in rData:
    row['Name']= name
    rows.append(row)

df= pd.DataFrame(rows)
df
```

```python
# using pivot_table
df = df.pivot_table(index='Name', columns=['Grade'],
					values=['Exam']).reset_index()
# Defining columns
df.columns = ['Name', 'Maths', 'Physics', 'Chemistry']

# print dataframe
print(df)
```
